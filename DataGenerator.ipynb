### Import Libs

```python
from numpy import where
from sklearn.datasets import make_blobs
from matplotlib import pyplot
import seaborn as sns
import pandas as pd
import random
```

```python
sns.set_theme()
```

### Generator Function

```python
def data_generator(name):
    # define dataset
    X, y = make_blobs(n_samples=1000,centers=2, center_box=(-10.0, 10.0), cluster_std=2)
    # summarize dataset shape
    print(X.shape, y.shape)
    
    data = {"Coord_X": X[:,0],"Coord_Y":X[:,1],"label":y }
    df = pd.DataFrame.from_dict(data)
    df["label"] = df.label.replace(0,"Blue")
    df["label"] = df.label.replace(1,"Red")
    
    pyplot.figure(figsize=(10,7))
    ax = sns.scatterplot(data=df,x=df.Coord_X,y=df.Coord_Y,hue=df.label)
    ax.set_title(name)
    pyplot.show()
    
    df.to_csv("./data/"+name+".csv",index=None)
```

### Generate Samples

```python
dataset = ["sample"+str(i) for i in range(1,11)]
for data in dataset:
    data_generator(data)
```

### Curated Samples

Select samples are placed in *data-acc* folder. Visual representation is as given below 

```python
dataset = ["sample"+str(i) for i in range(1,11)]
figure, axes = pyplot.subplots(5, 2)
figure.set_figwidth(20)
figure.set_figheight(26)
figure.subplots_adjust(hspace=0.3)
pyplot.title("Data Samples")
x = 0
y = 0
for i, data in enumerate(dataset):
    df = pd.read_csv("./data-acc/"+data+".csv")
    ax = sns.scatterplot(data=df,x=df.Coord_X,y=df.Coord_Y,hue=df.label,ax=axes[x,y])
    ax.set_title(data)
    
    if y == 0:
        y = 1
    else:
        y = 0
        x = x +1
pyplot.show()
figure.savefig("data_samples.jpg",dpi=300)
```
